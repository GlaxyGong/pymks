# Spinoidal Example for MKS

The following notebook describes the bone headed approach to MKS (without Fourier Transform) just with a least squares approach.

## Computing Influence Coefficients for a Spinoidal Decompostion

 - `N` - discretization of the spatial domain
 - `Nbins` - discretization of the local state space
 - `Nsamples` - number of samples in the regression

```python
import numpy as np
import fipy as fp
import pylab as plt

np.random.seed(3985)

N = 30
Nbins = 40
Nsamples = 50
dt = 1e-3
```

The following function takes a random field to seed one iteration of the spinoidal decomposition (i.e. calculate the response from a microstructure).

```python
def fipy_response(phi0, dt, N):
    
    nx = ny = N
    mesh = fp.PeriodicGrid2D(nx=nx, ny=ny, dx=0.25, dy=0.25)
    phi = fp.CellVariable(name=r"$\phi$", mesh=mesh, value=phi0.copy())
    PHI = phi.arithmeticFaceValue
    D = a = epsilon = 1.
    eq = (fp.TransientTerm()
      == fp.DiffusionTerm(coeff=D * a**2 * (1 - 6 * PHI * (1 - PHI)))
      - fp.DiffusionTerm(coeff=(D, epsilon**2)))
    
    eq.solve(phi, dt=dt, solver=fp.LinearLUSolver())
    
    return phi0, (np.array(phi) - phi0) / dt
    
samples = [fipy_response(np.random.random(N**2), dt=dt, N=N) for i in range(Nsamples)]
```

The function `bin`, discretizes the microstructure.

 * `ms` is the initial microstructure
 * `Nbins` is the number of bins
 

```python
def bin(ms, Nbins):
    H = np.linspace(0, 1, Nbins)
    dh = H[1] - H[0]
    return np.maximum(1 - abs(ms[:,np.newaxis] - H) / dh, 0)
    
microstructures = [bin(phi0, Nbins) for phi0, response in samples]
```

```python
fourierMS = np.fft.fft2( ms, axes=( 1, 2 ) ) ;
fourierResponse = np.fft.fft2( samples[ :, 1, :].reshape( (Nsamples, N, N) ), axes=( 1, 2 ) );
```

```python
InfluenceCoeff = np.zeros( (N, N, Nbins), dtype=np.complex);
```

```python
for ii in range(N):
    for jj in range(N):
        a = fourierMS[:,ii,jj,:].reshape((Nsamples, Nbins))
        InfluenceCoeff[ii,jj,:] = np.linalg.lstsq( a, fourierResponse[:,ii,jj] )[0]
        
```

```python
def microstructureMatrix(microstructure, N, Nbins):
    matrix = np.zeros((N**2, N**2 * Nbins))
    for i in range(N**2):
        matrix[i] = np.roll(microstructure, -i, axis=0).swapaxes(0,1).flatten()
        
    return matrix
mMatrices = [microstructureMatrix(m, N, Nbins) for m in microstructures]
```
