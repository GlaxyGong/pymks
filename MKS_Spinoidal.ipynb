# Spinoidal Example for MKS

The following notebook describes the bone headed approach to MKS (without Fourier Transform) just with a least squares approach.

## Computing Influence Coefficients for a Spinoidal Decompostion

 - `N` - discretization of the spatial domain
 - `Nbins` - discretization of the local state space
 - `Nsamples` - number of samples in the regression

```python
import numpy as np
import fipy as fp
import pylab as plt

np.random.seed(3985)

N = 20
Nbins = 7
Nsamples = 50
dt = 1e-2
```

The following function takes a random field to seed one iteration of the spinoidal decomposition (i.e. calculate the response from a microstructure).

```python
def fipy_response(phi0, dt, N):
    mesh = fp.PeriodicGrid2D(dx=dx, dy=dx, nx=N, ny=N)
    phi = fp.CellVariable(mesh=mesh, value=phi0.copy())
    e = fp.TransientTerm() == fp.DiffusionTerm(3 * phi**2 - 1) - fp.DiffusionTerm((1,1))
    e.solve(phi, dt=dt, solver=fp.LinearLUSolver())
    return phi0, (np.array(phi) - phi0) / dt

samples = [fipy_response(np.random.random(N**2), L=N, dt=dt, N=N) for i in range(Nsamples)]
```

```python
nx = ny = 100

mesh = fp.Grid2D(nx=nx, ny=ny, dx=0.25, dy=0.25)
phi = fp.CellVariable(name=r"$\phi$", mesh=mesh)
 
viewer = fp.Viewer(vars=(phi,), datamin=0., datamax=1.)

PHI = phi.arithmeticFaceValue
D = a = epsilon = 1.
eq = (fp.TransientTerm()
      == fp.DiffusionTerm(coeff=D * a**2 * (1 - 6 * PHI * (1 - PHI)))
      - fp.DiffusionTerm(coeff=(D, epsilon**2)))

dexp = -5
elapsed = 0.
duration = 0.1

while elapsed < duration:
    dt = min(100, np.exp(dexp))
    elapsed += dt
    dexp += 0.01
    eq.solve(phi, dt=dt, solver=fp.LinearLUSolver())
    if __name__ == "__main__":
        viewer.plot()
    elif (max(phi.globalValue) > 0.7) and (min(phi.globalValue) < 0.3) and elapsed > 10.:
        break
    print elapsed


```

```python
def bin(x, Nbins):
    space = np.linspace(0, 1, Nbins)
    dx = space[1] - space[0]
    return np.maximum(1 - abs(x[:,np.newaxis] - space) / dx, 0)
    
microstructures = [bin(phi0, Nbins) for phi0, p in samples]
```

```python
def microstructureMatrix(microstructure, N, Nbins):
    matrix = np.zeros((N**2, N**2 * Nbins))
    for i in range(N**2):
        matrix[i] = np.roll(microstructure, -i, axis=0).swapaxes(0,1).flatten()
        
    return matrix
mMatrices = [microstructureMatrix(m, N, Nbins) for m in microstructures]
```

```python
alpha = np.linalg.lstsq(mMatrix, response)[0]

print response
print np.dot(mMatrix, alpha)


print abs(np.dot(mMatrix, alpha) - response)
    
```
